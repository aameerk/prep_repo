```python
class BankAccount:
  """Production style bank account."""

  def __init__(self, owner: str, balance:float = 0.0):
    self.owner = owner
    self._balance = balance
    self._transactions: list = []

  def deposit(self, amount:float) -> None:
    if amount<=0:
      raise ValueError("Deposit must be positive")
    self._balance += amount
    self._transactions.append(("deposit",amount))
  def withdraw(self,amount:float) -> None:
    if amount> self._balance:
      raise ValueError("Insufficient Funds")
    self._balance -=amount
    self._transactions.append(("withdrawal",amount))

  def getbalance(self) -> float:
    return self._balance

acc = BankAccount("Alice",1000.0)
acc.deposit(500)
acc.withdraw(250)
acc.getbalance()
```

```python
import math

class Circle:
  PI = math.pi #class variable

  def __init__(self, radius: float):
    self.radius = radius

  def area(self) -> float:
    return Circle.PI * self.radius ** 2

  def circumference(self) -> float:
    return 2 * Circle.PI * self.radius

  @classmethod
  def from_diameter(cls , diameter: float) -> 'Circle':
    return cls(diameter/2)

  @staticmethod
  def is_valid_radius(value:float) -> bool:
    return value>0

c1 = Circle(5)
c2 = Circle.from_diameter(10)

print(c1.area())
print(c1.circumference())
print(c2.area())
print(c2.circumference())
print(c2.from_diameter(-1))
print(Circle.is_valid_radius(-1))
```

```python
class Temperature:
  """Stores temperature in Celsius, exposes Fahrenheit too."""

  def __init__(self, celsius: float):
    self.celsius = celsius #calls the setter below

  @property
  def celsius(self) -> float:
    return self._celsius

  @celsius.setter
  def celsius(self, value:float) -> None:
    if value < -273.15:
      raise ValueError(f"Temperature {value}C is below absolute zero")
    self._celsius = value
  @property
  def fahrenheit(self) -> float:
    return self.celsius * 9/5 + 32
  @fahrenheit.setter
  def fahrenheit(self, value:float) -> None:
    self.celsius = (value - 32) * 5/9

  @property
  def kelvin(self) -> float:
    return self._celsius + 273.15

  def __repr__(self) -> str:
    return f"Temperature({self._celsius:.2f}C)"

t = Temperature(1000)

print(t.fahrenheit)
print(t.celsius)
print(t.kelvin)
```

```python
from __future__ import annotations
from typing import Iterator

class Vector:
  """2D vector with full operator support."""

  def __init__(self, x, y) -> float:
    self.x , self.y = x , y

    #string representations

  def __repr__(self) -> str:
    return f"Vector ({self.x}, {self.y})"
  def __str__(self) -> str:
    return f"({self.x}, {self.y})"
#arithmetic
  def __add__(self, other: Vector) -> Vector:
    return Vector(self.x + other.x, self.y + other.y)

  def __sub__(self, other: Vector) -> Vector:
    return Vector(self.x - other.x, self.y - other.y)
  def __mul__(self, other:Vector) -> Vector:
    return Vector(self.x * other.x, self.y * other.y)
  def __truediv__(self, other:Vector) -> Vector:
    return Vector(self.x / other.x, self.y / other.y)

  def __len__(self) -> int:
    return int((self.x**2 + self.y**2) ** 0.5)
  def __hash__(self) -> int:
    return hash((self.x , self.y))

  #Iteration
  def __iter__(self) -> Iterator[float]:
    yield self.x
    yield self.y


v1 = Vector(3,4)
v2 = Vector(1,3)

x, y= v1
print(v1)


print(hash(v1))
print(len(v1))
```

```python
from abc import ABC, abstractmethod
from typing import Dict, Any

class DataStore(ABC):
    """Interface every storage backend must implement"""

    @abstractmethod
    def connect(self) -> None:
        pass

    @abstractmethod
    def disconnect(self) -> None:
        pass

    @abstractmethod
    def read(self, key: str) -> Any:
        pass

    def read_or_default(self, key: str, default: Any = None) -> Any:
        try:
            return self.read(key)
        except KeyError:
            return default


class InMemoryStore(DataStore):
    def __init__(self):
        self._data: Dict[str, Any] = {}

    def connect(self) -> None:
        print("In-memory store ready")

    def disconnect(self):
        self._data.clear()

    def read(self, key: str) -> Any:
        return self._data[key]

    def write(self, key: str, value: Any) -> None:
        self._data[key] = value


class RedisStore(DataStore):          # ✅ sibling class, not nested
    def connect(self):
        print("Connecting to Redis")

    def disconnect(self):
        print("Redis connection closed")

    def read(self, key: str) -> Any:
        return None

    def write(self, key: str, value: Any) -> None:
        pass


def cache_user(store: DataStore, uid: int, data: dict) -> None:   # ✅ standalone function
    store.write(f"user:{uid}", data)


# ── Usage ──────────────────────────────────────────
store = InMemoryStore()
store.connect()                            # In-memory store ready

cache_user(store, 43, {"name": "Alice"})   # writes 'user:43'

# ✅ read — call on the instance with correct string key
print(store.read("user:43"))               # {'name': 'Alice'}

# ✅ read_or_default — won't crash if key missing
print(store.read_or_default("user:99", "not found"))   # not found

store.disconnect()
```

```python
from __future__ import annotations
from dataclasses import dataclass, field
from abc import ABC, abstractmethod
from typing import List, Optional
from datetime import datetime
from enum import Enum


@dataclass(frozen=True)
class Money:
    amount: float
    currency: str = 'INR'

    def __add__(self, other: Money) -> Money:
        assert self.currency == other.currency
        return Money(self.amount + other.amount, self.currency)

    def __str__(self) -> str:
        return f'{self.currency} {self.amount:,.2f}'


class OrderStatus(Enum):
    PENDING = 'pending'
    CONFIRMED = 'confirmed'
    SHIPPED = 'shipped'
    DELIVERED = 'delivered'
    CANCELLED = 'cancelled'


class Discount(ABC):
    @abstractmethod
    def apply(self, price: Money) -> Money:
        pass


class PercentDiscount(Discount):
    def __init__(self, pct: float):
        self.pct = pct                           # ✅ Bug 1 fixed

    def apply(self, price: Money) -> Money:
        return Money(price.amount * (1 - self.pct / 100), price.currency)


class FixedDiscount(Discount):
    def __init__(self, amount: float):
        self.amount = amount

    def apply(self, price: Money) -> Money:
        return Money(max(0, price.amount - self.amount), price.currency)


@dataclass
class Product:
    name: str
    _price: Money
    stock: int = 0
    discount: Optional[Discount] = None

    @property
    def price(self) -> Money:
        if self.discount:
            return self.discount.apply(self._price)  # ✅ Bug 2 fixed — use _price
        return self._price

    @property
    def available(self) -> bool:
        return self.stock > 0


class Order:
    _counter: int = 0

    def __init__(self, customer: str):
        Order._counter += 1
        self.order_id = f'ORD-{Order._counter:05d}'
        self.customer = customer
        self.items: List[tuple] = []
        self.status = OrderStatus.PENDING
        self.created_at = datetime.utcnow()

    def add_item(self, product: Product, qty: int = 1) -> None:
        if not product.available:
            raise ValueError(f"{product.name} is out of stock")
        self.items.append((product, qty))

    @property
    def total(self) -> Money:
        total = Money(0, 'INR')                          # ✅ Bug 4 fixed
        for product, qty in self.items:
            total = total + Money(product.price.amount * qty, 'INR')
        return total                                     # ✅ Bug 3 fixed — outside loop

    def confirm(self) -> None:
        self.status = OrderStatus.CONFIRMED

    def __repr__(self) -> str:
        return (
            f'Order({self.order_id}, customer={self.customer!r}, '
            f'items={len(self.items)}, total={self.total}, '
            f'status={self.status.value})'
        )


# ── Usage ──────────────────────────────────────────
laptop = Product('MAC', Money(950000), stock=10, discount=PercentDiscount(5))
order = Order("Alice")
order.add_item(laptop, 1)
print(order)
```

```python
class StringFinding:
    def __init__(self, string):
        self.string = string
        self.pos = string.find("self")

    def result(self):
        if self.pos == -1:
            return "'self' was not found."
        return f"'self found at position {self.pos}"


obj = StringFinding("hello self world")
print(obj.result())
```
